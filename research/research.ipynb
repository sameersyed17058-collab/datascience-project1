```python
# Q2
# Given an array of positive integers, count the number of contiguous subarrays whose product is less than k.
# Input: nums = [10,5,2,6], k = 100
# Output: 8


def count(nums = [10, 5, 2, 6], k=100):
  for index in range(0, len(nums)):
    if nums[index] * nums[index + 1] < k:
      if nums[index] < nums[index + 1]:
        return nums[index] + nums[index + 1]
```
